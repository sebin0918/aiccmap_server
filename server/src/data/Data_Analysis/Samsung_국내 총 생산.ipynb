```python
import datetime
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
ticker = '005930.KS' # 삼성전자 주식 코드
```

```python
# 지표 데이터 가져오기
stock_data = pd.read_excel('index_Infomation_v0.2.xlsx')
stock_data['Date'] = pd.to_datetime(stock_data['Date'])
stock_data
```

```python
# 삼성 주식 데이터 가져오기

df = stock_data[['Date', 'samsung_Close', '국내 총 생산', '중앙은행 기준금리']]
df
```

```python
# 데이터 타입 날짜형식으로 변경

df['Date'] = pd.to_datetime(stock_data['Date'])
```

```python
# 삼성전자 주식 그래프 그리기

plt.figure(figsize = (20, 5))
plt.plot(df['Date'], df['samsung_Close'], label='Samsung')

plt.title('Samsung Stock Price')
plt.xlabel('Date')
plt.ylabel('Price (KRW)')
plt.legend()
plt.grid(True)

plt.show()
```

```python
# 국내 총 생산 (GDP) 그래프 그리기

plt.figure(figsize = (20, 5))
plt.plot(df['Date'], df['국내 총 생산'], label = 'KOR GDP', color = 'green')

plt.title('KOR GDP')
plt.xlabel("Date")
plt.ylabel("KOR GDP (KOR)")
plt.legend()
plt.grid(True)

plt.show()
```

```python
# 삼성전자 & 국내 총생산 그래프 그리기

plt.figure(figsize = (20, 5))
plt.plot(df['Date'], df['samsung_Close'], label = 'Samsung')
plt.plot(df['Date'], df['국내 총 생산'], label = 'KOR GDP', color = 'green')

plt.title('Samsung vs GDP')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)

plt.show()
```

```python
# 정규화

def normalize(df, column):
    return (df[column] - df[column].min()) / (df[column].max() - df[column].min())
```

```python
samsung_normal_df = normalize(df, 'samsung_Close')
```

```python
GDP_normal_df = normalize(df, '국내 총 생산')
```

```python
# 정규화 후 그래프

plt.figure(figsize = (20, 5))
plt.plot(df['Date'], samsung_normal_df, label = 'Samsung')
plt.plot(df['Date'], GDP_normal_df, label = 'GDP', color = 'green')

plt.title("Samsung vs GDP (KOR)")
plt.xlabel('Date')
plt.ylabel('GDP')
plt.legend()
plt.grid(True)

plt.show()
```

```python
# 변동률 차이 비교

# 일간 변동률 계산

daily_change_samsung_df = df['samsung_Close'].pct_change()
daily_change_GDP_df = df['국내 총 생산'].pct_change()
```

```python
# 그래프 화

plt.figure(figsize = (20, 5))
plt.plot(df['Date'], daily_change_samsung_df, label = 'Samsung')
plt.plot(df['Date'], daily_change_GDP_df, label = 'GDP', color = 'green')

plt.title('Samsung vs GDP Change Rate')
plt.xlabel('Date')
plt.ylabel('Change Rate')

plt.legend()
plt.grid(True)

plt.show()
```

```python
#산점도 플롯

plt.figure(figsize = (10, 10))
plt.scatter(df['samsung_Close'], df['국내 총 생산'], alpha = 0.05)

plt.title('Samsung vs GDP (KOR)')
plt.xlabel('Samsung stock price (KRW)')
plt.ylabel('GDP')
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Calculate the correlation
correlation = df['국내 총 생산'].corr(df['samsung_Close'])
print(f"Correlation between GDP and Samsung Stock Price: {correlation:.2f}")
```
