```python
import datetime
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
ticker = '005930.KS' # 삼성전자 주식 코드
```

```python
# 삼성전자 주식 데이터 가져오기 

samsung = yf.Ticker(ticker)
samsung_df = samsung.history(start='2014-01-01')
samsung_df.reset_index(inplace=True)

# 삼성전자 주식 그래프 그리기

plt.figure(figsize=(14, 7))
plt.plot(samsung_df.index, samsung_df['Close'], label='Samsung Electronics')
plt.title('Samsung Electronics Stock Price')
plt.xlabel('Date')
plt.ylabel('Closing Price (KRW)')
plt.legend()
plt.grid(True)

plt.show()
```

```python
# 경제지표 데이터 가져오기

stock_data = pd.read_excel('index_Infomation_0630.xlsx')
stock_data.head()
```

```python
# 환율 (dollar to won Close)

dtw_df = stock_data[['Date', 'dollar_to_won_Close']]

dtw_df
```

```python
# 데이터 타입 날짜형식으로 변경

dtw_df['Date'] = pd.to_datetime(stock_data['Date'])
```

```python
# 환율 그래프

plt.figure(figsize = (14, 7))
plt.plot(dtw_df['Date'], dtw_df['dollar_to_won_Close'], label = 'Exchange Ratio', color = 'gold')

plt.title('Dollar_to_Won (Excahnge Ratio)')
plt.xlabel('Date')
plt.ylabel('Closing Price ($/KRW)')
plt.legend()
plt.grid(True)

plt.show()
```

```python
# 삼성전자 주식과 환율 그래프
plt.figure(figsize=(14, 7))

plt.plot(samsung_df['Date'], samsung_df['Close'], label='Samsung')
plt.plot(dtw_df['Date'], dtw_df['dollar_to_won_Close'], label='Exchange Ratio', color='gold')

plt.title('Samsung vs Exchange Ratio ($/KRW)')
plt.xlabel('Date')
plt.ylabel('Closing Price (KRW)')
plt.legend()
plt.grid(True)

plt.show()
```

```python
# 0과 1 사이의 값으로 정규화
def normalize(df, column):
    return (df[column] - df[column].min()) / (df[column].max() - df[column].min())
```

```python
samsung_normal_df = normalize(samsung_df, 'Close')
```

```python
dtw_normal_df = normalize(dtw_df, 'dollar_to_won_Close')
```

```python
# 정규화 적용 삼성 주식과 환율 그래프

plt.figure(figsize = (14, 7))
plt.plot(samsung_df['Date'], samsung_normal_df, label = 'Samsung')
plt.plot(dtw_df['Date'], dtw_normal_df, label = 'Exchange Ratio', color = 'gold')

plt.title('Samsung vs Exchange Ratio ($/KRW)')
plt.xlabel('Date')
plt.ylabel('Closing Price (KRW)')
plt.legend()
plt.grid(True)

plt.show()
```

```python
# 변동률 차이 비교

#일간 변동률 계산 (판다스 함수 이용), ## shift(1) 이용하면 차이 계산도 가능

daily_change_samsung_df = samsung_df['Close'].pct_change()
daily_change_dtw_df = dtw_df['dollar_to_won_Close'].pct_change()
```

```python
# 그래프화

plt.figure(figsize = (14, 7))
plt.plot(samsung_df['Date'], daily_change_samsung_df, label = 'Samsung')
plt.plot(dtw_df['Date'], daily_change_dtw_df, label = 'Exchange Ratio', color = 'gold')

plt.title('Samsung vs Exchange Ratio Change Rate')
plt.xlabel('Date')
plt.ylabel('Change Rate')

plt.legend()
plt.grid(True)

plt.show()
```

```python
# 삼성 주식과 환율 데이터 병합

# Date를 기준으로 데이터 프레임 병합

samsung_df['Date'] = samsung_df['Date'].dt.tz_localize(None)

merge_df = pd.merge(samsung_df[['Date', 'Close']], dtw_df[['Date', 'dollar_to_won_Close']], on='Date', how='inner')
merge_df
```

```python
# Calculate the correlation
correlation = merge_df['Close'].corr(merge_df['dollar_to_won_Close'])
print(f"Correlation between Exchange Ratio and Samsung Stock Price: {correlation:.2f}")
```

```python
# 산점도 그래프

plt.figure(figsize = (14, 7))
plt.scatter(merge_df['Close'], merge_df['dollar_to_won_Close'], alpha = 0.2, label = 'Data Points')

plt.title('Samsung vs Exchange Ratio scatter')
plt.xlabel('Samsung stock price (KRW)')
plt.ylabel('Exchange Ratio ($/KRW)')

plt.legend()
plt.grid(True)

plt.show()
```
