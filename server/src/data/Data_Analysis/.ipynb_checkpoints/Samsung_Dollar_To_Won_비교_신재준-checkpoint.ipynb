```python
# 경고 메시지를 제어하고 관리하기 위해 사용하는 표준 라이브러리.
import warnings
```

```python
# apple_close 데이터프레임의 'Date' 열을 선택 / datetime 형식으로 변환, 변환된 값을 다시 'Date' 열에 할당
samsung_close.loc[:, 'Date'] = pd.to_datetime(samsung_close['Date'])

# 그래프의 크기를 설정
plt.figure(figsize=(20, 5))
plt.plot(samsung_close['Date'], samsung_close['samsung_Close'], label='Samsung Electornics', color='red')
plt.title('Samsung Electornics Stock Price')
plt.xlabel('Date')
plt.ylabel('Closing Price (Dollar)')

# 현재 활성화된 Axes 객체를 가져와서 변수 ax에 할당 / gca ='get current axes'의 약자로, 현재 그래프에 대한 축 객체를 반환
ax = plt.gca()
# x축의 주요 눈금 위치를 설정 / mdates.YearLocator(1)은 매년 1년마다 주요 눈금을 설정하도록 지정 / 즉, 각 연도의 시작 부분에 주요 눈금이 표시됨.
ax.xaxis.set_major_locator(mdates.YearLocator(1))
# x축의 날짜 레이블 형식을 설정, mdates.DateFormatter('%Y-%m')은 날짜 레이블을 '년-월' 형식으로 표시하도록 지정 / 2024-01'과 같은 형식으로 날짜가 표시됨.
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))

plt.legend()
plt.grid(True)
plt.show()
```

```python
# 0과 1 사이의 값으로 정규화
def normalize(df, column):
    return (df[column] - df[column].min()) / (df[column].max() - df[column].min())
```

```python
samsung_normal_df = normalize(samsung_close, 'samsung_Close')
```

```python
dollar_to_won_normal_df = normalize(dollar_to_won_close, 'dollar_to_won_Close')
```

```python
# 정규화 적용 삼성 주식과 환율 그래프

plt.figure(figsize = (20, 5))
plt.plot(samsung_close['Date'], samsung_normal_df, label = 'Samsung')
plt.plot(dollar_to_won_close['Date'], dollar_to_won_normal_df, label = 'Exchange Ratio', color = 'gold')

plt.title('Samsung vs Dollar To Won ($/KRW)')
plt.xlabel('Date')
plt.ylabel('Closing Price (KRW)')
plt.legend()
plt.grid(True)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# 사본 생성 후 Date 컬럼을 datetime 형식으로 변환
dollar_to_won_close_copy = dollar_to_won_close.copy()
samsung_close_copy = samsung_close.copy()

dollar_to_won_close_copy['Date'] = pd.to_datetime(dollar_to_won_close_copy['Date'])
samsung_close_copy['Date'] = pd.to_datetime(samsung_close_copy['Date'])

# 일일 변동률 계산
dollar_to_won_close_copy['Dollar To Won Change %'] = dollar_to_won_close_copy['dollar_to_won_Close'].pct_change() * 100
samsung_close_copy['Samsung Change %'] = samsung_close_copy['samsung_Close'].pct_change() * 100

# Date를 기준으로 데이터프레임 병합
merged_df = pd.merge(dollar_to_won_close_copy[['Date', 'Dollar To Won Change %']], samsung_close_copy[['Date', 'Samsung Change %']], on='Date', how='inner')

# 변동률 (일일 퍼센트 변동) 그래프 그리기
fig, ax1 = plt.subplots(figsize=(20, 7))

ax1.plot(merged_df['Date'], merged_df['Dollar To Won Change %'], label='Dollar To Won Price Change %', color='goldenrod')
ax1.set_xlabel('Date')
ax1.set_ylabel('Dollar To Won Price Change %', color='goldenrod')
ax1.tick_params(axis='y', labelcolor='goldenrod')

ax1.xaxis.set_major_locator(mdates.YearLocator(1))
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))

ax2 = ax1.twinx()
ax2.plot(merged_df['Date'], merged_df['Samsung Change %'], label='Samsung Stock Price Change %', color='red')
ax2.set_ylabel('Samsung Stock Price Change %', color='red')
ax2.tick_params(axis='y', labelcolor='red')

plt.title('Dollar To Won Price and Samsung Electronics Stock Price Volatility')
fig.tight_layout()
plt.grid(True)

fig.legend(loc="upper left", bbox_to_anchor=(0.1,0.9), bbox_transform=ax1.transAxes)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 사본 생성 후 Date 컬럼을 datetime 형식으로 변환
dollar_to_won_close_copy = dollar_to_won_close.copy()
samsung_close_copy = samsung_close.copy()

dollar_to_won_close_copy['Date'] = pd.to_datetime(dollar_to_won_close_copy['Date'])
samsung_close_copy['Date'] = pd.to_datetime(samsung_close_copy['Date'])

# Date를 기준으로 데이터프레임 병합
merged_df = pd.merge(dollar_to_won_close_copy[['Date', 'dollar_to_won_Close']], samsung_close_copy[['Date', 'samsung_Close']], on='Date', how='inner')

# 산점도 그래프 
plt.figure(figsize=(10, 7))
plt.scatter(merged_df['dollar_to_won_Close'], merged_df['samsung_Close'], alpha=0.1)
plt.title('Correlation between Dollar To Won Price and Samsung Electronics Stock Price', color='black')
plt.xlabel('Dollar To Won Price (USD)', color='black')
plt.ylabel('Samsung Stock Price (USD)', color='black')
plt.grid(True)
plt.xticks(color='black')
plt.yticks(color='black')
plt.gca().set_facecolor('white')  # 배경색 검정
plt.show()
```
