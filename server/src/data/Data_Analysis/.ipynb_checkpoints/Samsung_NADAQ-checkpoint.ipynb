```python
import datetime
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
stock_data = pd.read_excel('../index_Infomation_v0.1.xlsx')
stock_data = stock_data[['Date', 'samsung_Close', 'nasdaq_Close']]
stock_data.loc[:, 'Date'] = pd.to_datetime(stock_data['Date'])
stock_data.head()
```

```python
plt.figure(figsize=(20, 5))
plt.plot(stock_data['Date'], stock_data['samsung_Close'], label='SAMSUNG Price', color='blue')
plt.title('Samsung Stock Price')
plt.xlabel('Date')
plt.ylabel('Closing Price (won)')

ax = plt.gca()
ax.xaxis.set_major_locator(mdates.YearLocator(1))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(20,5))
plt.plot(stock_data['Date'], stock_data['nasdaq_Close'], label='NASDAQ_Price', color='yellow')
plt.title('NASDAQ_Price')
plt.xlabel('Date')
plt.ylabel('Closing Price ($)')

ax = plt.gca()
ax.xaxis.set_major_locator(mdates.YearLocator(1))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 0과 1사이의 값으로 정규화
def normalize(df, column):
    return (df[column] - df[column].min()) / (df[column].max() - df[column].min())

stock_data['samsung_Close'] = normalize(stock_data, 'samsung_Close')
stock_data['nasdaq_Close'] = normalize(stock_data, 'nasdaq_Close')
```

```python
# 정규화 적용 두 그래프 합치기
plt.figure(figsize = (20, 5))
plt.plot(stock_data['Date'], stock_data['samsung_Close'], label = 'SAMSUNG', color='blue')
plt.plot(stock_data['Date'], stock_data['nasdaq_Close'], label = 'NASDAQ', color = 'yellow')

ax = plt.gca()
ax.xaxis.set_major_locator(mdates.YearLocator(1))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))

plt.title('Compare Samsung, NASDAQ')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate the correlation
correlation = stock_data['samsung_Close'].corr(stock_data['nasdaq_Close'])
print(f"Correlation between Exchange Ratio and Samsung Stock Price: {correlation:.2f}")
```

```python
fluctuation_samsung = stock_data['samsung_Close'].pct_change()
fluctuation_nasdaq = stock_data['nasdaq_Close'].pct_change()
```

```python
plt.figure(figsize = (20, 5))
plt.plot(stock_data['Date'], fluctuation_samsung, label = 'SAMSUNG', color='blue')
plt.plot(stock_data['Date'], fluctuation_nasdaq, label = 'NASDAQ', color='yellow')
plt.ylim(-1, 1)
plt.title('SAMSUNG, NASDQ Raito fluctuation')
plt.xlabel('Date')
plt.ylabel('Change Rate')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# 가상의 데이터 예시
# stock_data = pd.read_csv('your_data.csv')

# 정규화된 주가 데이터
stock_data['norm_samsung_Close'] = stock_data['samsung_Close'] / stock_data['samsung_Close'].iloc[0]
stock_data['norm_nasdaq_Close'] = stock_data['nasdaq_Close'] / stock_data['nasdaq_Close'].iloc[0]

# 변동률 계산
fluctuation_samsung = stock_data['samsung_Close'].pct_change()
fluctuation_nasdaq = stock_data['nasdaq_Close'].pct_change()

# 플롯 그리기
fig, ax1 = plt.subplots(figsize=(20, 5))

# 첫 번째 y축 (정규화된 주가)
ax1.plot(stock_data['Date'], stock_data['norm_samsung_Close'], label='SAMSUNG (Normalized)', color='blue')
ax1.plot(stock_data['Date'], stock_data['norm_nasdaq_Close'], label='NASDAQ (Normalized)', color='yellow')
ax1.set_xlabel('Date')
ax1.set_ylabel('Normalized Closing Price')
ax1.set_title('Compare SAMSUNG and NASDAQ with Rate Fluctuation')
ax1.xaxis.set_major_locator(mdates.YearLocator(1))
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
ax1.legend(loc='upper left')
ax1.grid(True)

# 두 번째 y축 (변동률)
ax2 = ax1.twinx()
ax2.plot(stock_data['Date'], fluctuation_samsung, label='SAMSUNG Rate Change', color='blue', alpha=0.3)
ax2.plot(stock_data['Date'], fluctuation_nasdaq, label='NASDAQ Rate Change', color='green', alpha=0.3)
ax2.set_ylabel('Change Rate')
ax2.set_ylim(-1, 1)
ax2.legend(loc='upper right')

plt.show()
```

```python
# 산점도 그래프

plt.figure(figsize = (14, 7))
plt.scatter(stock_data['samsung_Close'], stock_data['nasdaq_Close'], alpha = 0.05, label = 'Data Points')

plt.title('SAMSUNG vs NASDAQ')
plt.xlabel('SAMSUNG')
plt.ylabel('NASDAQ')

plt.legend()
plt.grid(True)

plt.show()
```
