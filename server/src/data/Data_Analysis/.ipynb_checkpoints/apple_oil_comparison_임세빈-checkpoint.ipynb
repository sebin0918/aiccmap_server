```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# 사본 생성 후 Date 컬럼을 datetime 형식으로 변환
oil_close_copy = oil_close.copy()
apple_close_copy = apple_close.copy()

oil_close_copy['Date'] = pd.to_datetime(oil_close_copy['Date'])
apple_close_copy['Date'] = pd.to_datetime(apple_close_copy['Date'])

# 일일 변동률 계산
oil_close_copy['Oil Change %'] = oil_close_copy['oil_Close'].pct_change() * 100
apple_close_copy['Apple Change %'] = apple_close_copy['apple_Close'].pct_change() * 100

# Date를 기준으로 데이터프레임 병합
merged_df = pd.merge(oil_close_copy[['Date', 'Oil Change %']], apple_close_copy[['Date', 'Apple Change %']], on='Date', how='inner')

# 변동률 (일일 퍼센트 변동) 그래프 그리기
fig, ax1 = plt.subplots(figsize=(20, 7))

ax1.plot(merged_df['Date'], merged_df['Oil Change %'], label='Oil Price Change %', color='goldenrod')
ax1.set_xlabel('Date')
ax1.set_ylabel('Oil Price Change %', color='goldenrod')
ax1.tick_params(axis='y', labelcolor='goldenrod')

ax1.xaxis.set_major_locator(mdates.YearLocator(1))
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))

ax2 = ax1.twinx()
ax2.plot(merged_df['Date'], merged_df['Apple Change %'], label='Apple Stock Price Change %', color='red')
ax2.set_ylabel('Apple Stock Price Change %', color='red')
ax2.tick_params(axis='y', labelcolor='red')

plt.title('Oil Price and Apple Electronics Stock Price Volatility')
fig.tight_layout()
plt.grid(True)

fig.legend(loc="upper left", bbox_to_anchor=(0.1,0.9), bbox_transform=ax1.transAxes)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 사본 생성 후 Date 컬럼을 datetime 형식으로 변환
oil_close_copy = oil_close.copy()
apple_close_copy = apple_close.copy()

oil_close_copy['Date'] = pd.to_datetime(oil_close_copy['Date'])
apple_close_copy['Date'] = pd.to_datetime(apple_close_copy['Date'])

# Date를 기준으로 데이터프레임 병합
merged_df = pd.merge(oil_close_copy[['Date', 'oil_Close']], apple_close_copy[['Date', 'apple_Close']], on='Date', how='inner')

# 산점도 그래프 
plt.figure(figsize=(10, 7))
plt.scatter(merged_df['oil_Close'], merged_df['apple_Close'], alpha=0.2)
plt.title('Correlation between Oil Price and Apple Electronics Stock Price', color='black')
plt.xlabel('Oil Price (USD)', color='black')
plt.ylabel('Apple Stock Price (USD)', color='black')
plt.grid(True)
plt.xticks(color='black')
plt.yticks(color='black')
plt.gca().set_facecolor('white')  # 배경색 흰색
plt.show()
```

```python
import pandas as pd
import numpy as np

oil_close_copy = oil_close.copy()
apple_close_copy = apple_close.copy()
oil_close_copy['Date'] = pd.to_datetime(oil_close_copy['Date'])
apple_close_copy['Date'] = pd.to_datetime(apple_close_copy['Date'])

# 0 값을 가지는 행을 제외한 후 일일 변동률 계산
oil_close_copy = oil_close_copy[oil_close_copy['oil_Close'] != 0]
apple_close_copy = apple_close_copy[apple_close_copy['apple_Close'] != 0]

oil_close_copy['Oil Change %'] = oil_close_copy['oil_Close'].pct_change() * 100
apple_close_copy['Apple Change %'] = apple_close_copy['apple_Close'].pct_change() * 100

# Date를 기준으로 데이터프레임 병합
merged_df = pd.merge(oil_close_copy[['Date', 'Oil Change %']], apple_close_copy[['Date', 'Apple Change %']], on='Date', how='inner')

# 결측치 및 무한대 값 제거
merged_df = merged_df.replace([np.inf, -np.inf], np.nan).dropna()

# 데이터의 기본 통계 출력
print(merged_df.describe())

# 결측치 확인
print(merged_df.isna().sum())

# 애플 주식과 유가의 상관계수 계산
correlation = merged_df['Oil Change %'].corr(merged_df['Apple Change %'])
print(f"Correlation between Oil Price and Apple Stock Price: {correlation:.2f}")
```
